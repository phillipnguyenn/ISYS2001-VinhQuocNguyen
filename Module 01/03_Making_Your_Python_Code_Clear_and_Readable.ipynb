## Writing Code Someone Else Can Read

Getting Python to run is only half the job. The other half is making your code — and your output — something a human can actually follow. A wall of unbroken text is just as hard to read as a wall of unbroken code.

In this notebook you'll pick up three small habits:

- Using blank lines to give your output room to breathe
- Printing multiple lines from a single `print()` call
- Writing comments that explain *why*, not just *what*

None of this changes what your program does. It changes whether the next person reading it — including you, in six months — can make sense of it in ten seconds instead of ten minutes.

Use your AI assistant as much as you like in this notebook, but the real goal isn't just finishing each exercise — it's understanding how `print()` actually behaves. That understanding is what lets you tell whether an AI's suggestion is actually good, or just confident-sounding.

## Section 10: A Blank Line in Your Code Does Nothing

An empty line *in your code* has no effect on the output — Python just skips straight over it. For example:

```python
print("Apple")

print("or Orange?")
```

**Output:**
```
Apple
or Orange?
```

Guess what the cell below will print, then run it and check.


````python
print("How")

print("are")
print("you?")
````

> *Hint:* Every `print()` starts a new line, whether or not there's a blank line in your code above it.

## Section 11: A Blank Line in Your Output Takes Work

So how do you get an actual blank line to show up in the output? Call `print()` with nothing inside it:

```python
print("How")
print()
print("are")
print("you?")
```

**Output:**
```
How

are
you?
```

Notice that second, empty line — that's what `print()` on its own does.

## Section 12: Multiple Lines, One Function

Can a single `print()` call produce several lines? Yes — two tricks do this.

**Trick one: `\n` starts a new line**

```python
print("This is the first line.\nThis is the second line.")
```

**Output:**
```
This is the first line.
This is the second line.
```

`\n` inside a string tells Python "jump to a new line here and keep printing." Simple, and useful when the whole thing is one string.

> *Tip:* Reach for `\n` when you're building a message inside a single string.

## Section 13: Triple Quotes

Python has a second trick for multi-line output: wrap your text in three single quotes (`'''`) or three double quotes (`"""`).

```python
print('''My contact:
Email: your_name@gmail.com
Phone: +xx-1234-5678''')
```

**Output:**
```
My contact:
Email: your_name@gmail.com
Phone: +xx-1234-5678
```

> *Advanced note:* Triple quotes preserve the line breaks exactly as you typed them — handy for anything with a fixed multi-line layout.

## Section 17: Comments

One last thing before you go: the line of code the computer *never* runs.

A comment doesn't affect your program at all, but it explains what's going on to any human reading it — including future-you. In Python, anything after a `#` on a line is ignored by the computer.

```python
# This line is a comment and won't be executed
print("Do you have any comment?")
```

**Output:**
```
Do you have any comment?
```
